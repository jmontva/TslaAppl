```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
Apple_csv_path = Path("Resources/HistoricalQuotes1.csv")
Tesla_csv_path = Path("Resources/Tesla.csv - Tesla.csv.csv")
```

```python
Apple_df = pd.read_csv(Apple_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
Tesla_df = pd.read_csv(Tesla_csv_path, index_col="Date", parse_dates=True, infer_datetime_format=True)
```

```python
combined_df = pd.concat([Apple_df, Tesla_df], axis="columns", join="inner")
combined_df.sort_index(inplace=True)
combined_df.head()
```

```python
combined_df.drop(columns=['Volume', 'Open', 'High', 'Low'], inplace=True)
combined_df.head()
```

```python
daily_returns = combined_df.pct_change()
daily_returns.head()
```
